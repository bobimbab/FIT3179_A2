```python
import pandas as pd
# Subnational 1 tree cover loss
rw = pd.read_csv("Subnational 1 tree cover loss.csv")

# Reshape the DataFrame
id_vars = ['country', 'subnational1', 'threshold', 'area_ha', 'extent_2000_ha', 'extent_2010_ha', 'gain_2000-2020_ha']
value_vars = [col for col in rw.columns if 'tc_loss_ha_' in col]
df_long = pd.melt(rw, id_vars=id_vars, value_vars=value_vars, var_name='year', value_name='tree_cover_loss')

# Convert year from 'tc_loss_ha_YYYY' to 'YYYY'
df_long['year'] = df_long['year'].str.replace('tc_loss_ha_', '').astype(int)

# Specify the new order of columns, moving 'year' to the third position
new_column_order = ['country', 'subnational1', 'year'] + [col for col in df_long.columns if col not in ['country','subnational1', 'year']]

# Reorder the columns
df_long = df_long[new_column_order]

# Display the reshaped DataFrame
print(df_long.head())

df_long.to_csv("Cleaned Subnational 1 tree cover loss.csv", index=False)
```
